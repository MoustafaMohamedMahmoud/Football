```python
"""
We Will create a function to print the bar chart for any stat and any position. This will make it easier to visualize the top players in different stats and positions.
And It Will Make The Code More Reusable and Cleaner.
"""
def Print_Player_Bar_Chart_Best(stat, position):
    fig, ax = plt.subplots(figsize=(10, 6))
    fig.set_facecolor('white')

    top10 = position.nlargest(10, stat)[['Player', stat]].reset_index(drop=True)

    colors = plt.cm.RdYlGn(np.linspace(1, 0, len(top10)))
    bars = ax.barh(top10['Player'], top10[stat], color=colors)

    for bar, val in zip(bars, top10[stat]):
         ax.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height() / 2,
            f'{val:.2f}', va='center', ha='left', fontsize=9, color='black')

    ax.set_title(f'{stat} — Best', color='black', fontsize=12, fontweight='bold')
    ax.set_facecolor('white')
    ax.tick_params(colors='black')
    ax.spines['bottom'].set_color('#ccc')
    ax.spines['left'].set_color('#ccc')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.tight_layout()
    #plt.savefig(f'{stat}.png', dpi=150, bbox_inches='tight', facecolor='white')
    plt.show()
```

```python
def Print_Player_Bar_Chart_Inverse(stat, position):
    fig, ax = plt.subplots(figsize=(10, 6))
    fig.set_facecolor('white')

    top10 = position.nsmallest(10, stat)[['Player', stat]].reset_index(drop=True)

    colors = plt.cm.RdYlGn(np.linspace(1, 0, len(top10)))
    bars = ax.barh(top10['Player'], top10[stat], color=colors)

    for bar, val in zip(bars, top10[stat]):
         ax.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height() / 2,
            f'{val:.2f}', va='center', ha='left', fontsize=9, color='black')

    ax.set_title(f'{stat} — Inverse', color='black', fontsize=12, fontweight='bold')
    ax.set_facecolor('white')
    ax.tick_params(colors='black')
    ax.spines['bottom'].set_color('#ccc')
    ax.spines['left'].set_color('#ccc')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.tight_layout()
    #plt.savefig(f'{stat}.png', dpi=150, bbox_inches='tight', facecolor='white')
    plt.show()
```
