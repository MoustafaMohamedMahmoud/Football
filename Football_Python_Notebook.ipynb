```python
wingers['Carries_To_Pen_p90'] = wingers['CPA'] / wingers['90s']
wingers['Carries_To_Pen_p90']
```

```python
def get_colors(values):
    normalized = (values - values.min()) / (values.max() - values.min())
    colors = []
    for v in normalized:
        if v > 0.66: colors.append('#2ECC71')
        elif v > 0.33: colors.append('#F39C12')
        else: colors.append('#E74C3C')
    return colors

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.patch.set_facecolor('white')

metrics = [
    ('Prog_Carries_p90', 'Progressive Carries per 90'),
    ('Non_Prog_Carries_p90', 'Non-Progressive Carries per 90'),
    ('Carries_To_Pen_p90', 'Carries to Penalty Area per 90')
]

for ax, (col, title) in zip(axes, metrics):
    top10 = wingers.nlargest(10, col)[['Player', col]].reset_index(drop=True)
    colors = get_colors(top10[col])
    bars = ax.barh(top10['Player'], top10[col], color=colors)
    
    for bar, val in zip(bars, top10[col]):
        ax.text(bar.get_width() + 0.05, bar.get_y() + bar.get_height()/2,
                f'{val:.2f}', va='center', ha='left', fontsize=9, color='black')
    
    ax.set_title(title, color='black', fontsize=11, fontweight='bold')
    ax.set_facecolor('white')
    ax.tick_params(colors='black')
    ax.spines['bottom'].set_color('#ccc')
    ax.spines['left'].set_color('#ccc')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('carrying_analysis.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
```
